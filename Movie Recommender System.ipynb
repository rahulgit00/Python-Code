```python
import pandas as pd
```

```python
df = pd.read_csv("./movie.csv", sep = "?")
titles = pd.read_csv("./Movie_Id_Titles.csv")
```

```python
df.head()
```

```python
data = pd.merge(df, titles, on = "item_id")
```

```python
data.head()
```

```python
g = data.groupby("title")["rating"].mean().sort_values(ascending = False).head()
g
```

```python
g = data.groupby("title")["rating"].count().sort_values(ascending = False).head()
g
```

```python
rating = pd.DataFrame(data.groupby("title")["rating"].mean())
```

```python
rating["Number of Rating"] = data.groupby("title")["rating"].count()
```

```python
rating.head()
```

```python
data.shape
```

```python
movie = data.pivot_table(index = 'user_id', columns='title', values="rating")
```

```python
movie.head()
```

```python
starwars_user_rating = movie["Star Wars (1977)"]
```

```python
starwars_user_rating.head()
```

```python
similar_to_starwars = movie.corrwith(starwars_user_rating)
```

```python
corr_starwars2 = pd.DataFrame(similar_to_starwars, columns=["Correlation"])
corr_starwars2.dropna(inplace = True)
corr_starwars2.head()
```

```python
corr_starwars2.sort_values("Correlation", ascending=False)
#corr_starwars2 = corr_starwars2.join(rating["Number of Rating"])


movies_100 = corr_starwars2[corr_starwars2["Number of Rating"] > 100]
movies_100.sort_values("Correlation", ascending = False).head()
```
